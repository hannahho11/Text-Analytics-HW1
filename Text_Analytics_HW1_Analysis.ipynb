```python
import numpy as np
import pandas as pd

df = pd.read_csv('Edmunds_Forums_Comments.csv',sep='\t')
df = df.drop('Unnamed: 0', axis=1)
df.head()
```

```python
import nltk
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english')) 
```

```python
counts = dict()

for i in range(0, len(df['comment'])):
    line = df['comment'][i]
    words = line.split()

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            #if not word in stop_words:
            counts[word] = 1
```
