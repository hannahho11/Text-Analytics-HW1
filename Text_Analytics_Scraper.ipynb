```
import pandas as pd

driver = webdriver.Chrome('chromedriver', options=chrome_options)
```

```
# Source: https://towardsdatascience.com/web-scraping-using-selenium-python-8a60f4cf40ab

# date, userid, and message 
userid = []
date = []
comment = []

driver.get('https://forums.edmunds.com/discussion/2864/general/x/entry-level-luxury-performance-sedans')

for i in range(150, 300):
  driver.get('https://forums.edmunds.com/discussion/2864/general/x/entry-level-luxury-performance-sedans' + str(i))
  for j in range(1,50):
    userid_element = driver.find_elements_by_xpath('/html/body/div[1]/main/div/div[3]/div[5]/div[1]/ul/li[' + str(j) + ']/div/div[2]/div[1]/span[1]/a[2]')[0]
    userid.append(userid_element.text)
    user_date = driver.find_elements_by_xpath('/html/body/div[1]/main/div/div[3]/div[5]/div[1]/ul/li[' + str(j) + ']/div/div[2]/div[2]/span/a/time')[0]
    date.append(user_date.get_attribute('title'))
    user_message = driver.find_elements_by_xpath('/html/body/div[1]/main/div/div[3]/div[5]/div[1]/ul/li[' + str(j) + ']/div/div[3]/div/div[1]')[0]
    comment.append(user_message.text)




/html/body/div[1]/main/div/div[3]/div[3]/div[2]/div/div[1]/div[1]/span[1]/a[2]
```

```
df = pd.DataFrame(list(zip(userid, date, comment)), columns=['username', 'date', 'comment']) 

df.head()
```
